```python
import pandas as pd
import matplotlib.pyplot as pl
import numpy as np
```

```python
df = pd.read_csv('percentagesReviewsGame.csv')
df.head()
```

```python
df_bar = df.head(25)
fig = df_bar.plot(x='app_name', y='porcentaje', kind='bar')
df_bar1 = df.tail(25)
fig1 = df_bar1.plot(x='app_name', y='porcentaje', kind='bar')
```

```python
bar = fig.get_figure()
bar.savefig('bestPorcentajesReviewsBar')
bar1 = fig1.get_figure()
bar1.savefig('worstPorcentajesReviewsBar')
```

```python
df1 = pd.read_csv('numReviewsForGame.csv')
lessReviews = df1.sort_values(by='count', ascending=True).head(25)
```

```python
lessReviews.plot.bar(x='app_name').get_figure().savefig('lessReviews')
```

```python
moreReviews = df1.sort_values(by='count', ascending=False).head(25)
moreReviews.plot.bar(x='app_name').get_figure().savefig('moreReviews')
```

```python
df2 = pd.read_csv('averageWordsForGame.csv')
lessWordsReviews = df2.sort_values(by='avg(len_review)', ascending=True).head(25)
```

```python
lessWordsReviews.plot.bar(x='app_name').get_figure().savefig('lessWordsReviews')
```

```python
df2 = pd.read_csv('averageWordsForGame.csv')
moreWordsReviews = df2.sort_values(by='avg(len_review)', ascending=False).head(25)
```

```python
moreWordsReviews.plot.bar(x='app_name').get_figure().savefig('moreWordsReviews')
```

```python
union = pd.merge(left=df2,right=df1, left_on='app_name', right_on='app_name')
union.head()
```

```python
union.plot.scatter(x='avg(len_review)', y='count')
```

```python
x=list(union['count'])
y=list(union['avg(len_review)'])
np.corrcoef(x, y)
```
